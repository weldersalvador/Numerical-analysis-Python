```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(t,y):
    return -2*y
```

```python
def explicit_euler_method(y0,l,u,function,h):
    N = int(np.ceil((u - l) / h))
    y = np.zeros(N + 1)
    t = np.zeros(N + 1)
    y[0] = y0
    t[0] = l
    for i in range(N):
        y[i + 1] = y[i] + function(t[i],y[i])*h
        t[i + 1] = l + i*h
    
    return t,y;
    
    
```

Specific for f(t,y) = -2*y

```python
def implicit_euler_method(y0,l,u,function,h):
    N = int(np.ceil((u - l) / h))
    y = np.zeros(N + 1)
    t = np.zeros(N + 1)
    y[0] = y0
    t[0] = l
    for i in range(N):
        y[i + 1] = y[i]/(2*h + 1)
        t[i + 1] = t[i] + h
        
    return t,y
```

```python
h = np.array([0.05,0.25,0.5,0.75,1])
```

```python
def plot_aprox(y0,l,u,function,h,method):
    for step in h:
        x,y = method(y0,l,u,function,step)
        plt.plot(x,y,label = f"h = {step}")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend(bbox_to_anchor=(1.05, 0.5))
    plt.grid()
    if (method == explicit_euler_method):
        plt.title("explicit_euler_method")
        plt.savefig("Explicit euler method")
    elif (method == implicit_euler_method):
        plt.title("implicit_euler_method")
        plt.savefig("Implicit euler method")
    plt.show()


plot_aprox(1,0,3,func,h = h, method = explicit_euler_method)
            
```

```python
plot_aprox(1,0,3,func,h = h, method = implicit_euler_method)
```

```python
def plot_real(l,u,function):
    x = np.linspace(l,u)
    y_real = function(x)
    plt.plot(x,y_real)
    plt.grid()
    plt.show()
```
